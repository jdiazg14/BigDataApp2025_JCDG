#3.ETL (Transformación y limpieza de datos)
---
Limpieza y transformación de los datos:
1. Borrado de registris con datos nulos
2. Crear nuevas columnas si son necesarias
3. Transformar columnas en datos reconocidos (números/texto/fechas/listas)

```python
# Eliminar filas con valores nulos (si existieran)
#df_bogota.dropna(inplace=True)
#df_palmira.dropna(inplace=True)
```

```python
# Quitar la coma (,) de la columna VALOR_MULTA
df_bogota["VALOR_MULTA"] = df_bogota["VALOR_MULTA"].astype(str).str.replace(',','',regex=False)
df_bogota["VALOR_MULTA"] = pd.to_numeric(df_bogota["VALOR_MULTA"], errors='coerce')
print (df_bogota.info())
```

```python
# Quitar la coma (,) de la columna VALOR_MULTA en Palmira
df_palmira["VALOR_MULTA"] = df_palmira["VALOR_MULTA"].astype(str).str.replace(',','',regex=False)
df_palmira["VALOR_MULTA"] = pd.to_numeric(df_palmira["VALOR_MULTA"], errors='coerce')
print (df_palmira.info())
```

```python
# Crear una columna con un valor numérico si la multa fue pagada o no
df_bogota["PAGADO_SI_NO_NUM"] = df_bogota["PAGADO_SI_NO"].apply(lambda x: 1 if x == 'SI' else 0)
df_palmira["PAGADO_SI_NO_NUM"] = df_palmira["PAGADO_SI_NO"].apply(lambda x: 1 if x == 'SI' else 0)
print (df_bogota.info())
```

#4.Agregación y estadísticas descriptivas
---
- Filtros/acumulado
- Funciones básicas de Pandas

```python
from inspect import BoundArguments
# Estadísticas descriptivas con VALOR_MULTA
with pd.option_context('display.float_format', '{:f}'.format):
    print(df_bogota["VALOR_MULTA"].describe())

# Calcular el promedio de multas por vigencia (año)
multas_promedio_por_vigencia_bogota = df_bogota.groupby('VIGENCIA')['VALOR_MULTA'].mean()
print(multas_promedio_por_vigencia_bogota)
```

```python
# Estadísticas descriptivas con VALOR_MULTA
with pd.option_context('display.float_format', '{:f}'.format):
    print(df_palmira["VALOR_MULTA"].describe())

# Calcular el promedio de multas por vigencia (año)
multas_promedio_por_vigencia_palmira = df_palmira.groupby('VIGENCIA')['VALOR_MULTA'].mean()
print(f"\nMultas por vigencia en Palmira: {multas_promedio_por_vigencia_palmira}")
```

#5.Análisis Avanzados

```python
# Tabla para ver el porcentaje de multas pagadas (si/no) por vigencia
multas_counts_bogota = df_bogota.groupby(['VIGENCIA', 'PAGADO_SI_NO']).size().unstack(fill_value=0)
multas_counts_bogota['TOTAL'] = multas_counts_bogota.sum(axis=1)
multas_porcentaje_bogota = multas_counts_bogota.div(multas_counts_bogota['TOTAL'], axis=0) * 100
print("Porcentaje de multas pagadas por vigencia(año)")
print(multas_porcentaje_bogota)
```

```python
# Tabla para ver el porcentaje de multas pagadas (si/no) por vigencia en Palmira
multas_counts_palmira = df_palmira.groupby(['VIGENCIA', 'PAGADO_SI_NO']).size().unstack(fill_value=0)
multas_counts_palmira['TOTAL'] = multas_counts_palmira.sum(axis=1)
multas_porcentaje_palmira = multas_counts_palmira.div(multas_counts_palmira['TOTAL'], axis=0) * 100
print("Porcentaje de multas pagadas por vigencia(año)")
print(multas_porcentaje_palmira)
```
